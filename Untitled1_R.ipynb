```r
library(data.table)     # For efficient data manipulation
library(dplyr)          # For data transformation
library(ggplot2)        # For visualizations
library(tidyr)          # For reshaping data
library(corrplot)       # For correlation plots
library(GGally)
```

```r
# load the dataset
df <- read.csv("diabetes.csv")

head(df)
```

```r
cols_with_missing <- c("Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI")

# Replace zeros with NA
df <- df |>
  mutate(across(all_of(cols_with_missing), ~ ifelse(. == 0, NA, .)))
```

```r
sum(is.na(df))
```

```r
# Impute missing values with median
df <- df |>
  mutate(across(all_of(cols_with_missing), ~ ifelse(is.na(.), median(., na.rm = TRUE), .)))
```

```r
# Verify no missing values remain
sum(is.na(df))
```

```r
# Function to identify outliers using IQR
identify_outliers <- function(column) {
  Q1 <- quantile(column, 0.25, na.rm = TRUE)
  Q3 <- quantile(column, 0.75, na.rm = TRUE)
  IQR <- Q3 - Q1
  lower_bound <- Q1 - 1.5 * IQR
  upper_bound <- Q3 + 1.5 * IQR
  return(column < lower_bound | column > upper_bound)
}
```

```r
# Apply the function to identify outliers
outliers <- df |>
  select(-Outcome) |>  # Exclude the target column
  mutate(across(everything(), identify_outliers))

# Check the number of outliers per column
colSums(outliers)
```

```r
# Create a mask to filter rows with no outliers
cleaned_df <- df |>
  mutate(is_outlier = rowSums(outliers) > 0) |>
  filter(!is_outlier) |>
  select(-is_outlier)

# Compare the dimensions of the original and cleaned datasets
cat("Original dataset dimensions: ", dim(df), "\n")
cat("Cleaned dataset dimensions: ", dim(cleaned_df), "\n")
```

```r
cleaned_df |>
  pivot_longer(-Outcome, names_to = "Variable", values_to = "Value") |>
  ggplot(aes(x = Value, fill = as.factor(Outcome))) +
  geom_density(alpha = 0.6) +
  facet_wrap(~Variable, scales = "free") +
  labs(fill = "Outcome", title = "Density Plots by Outcome") +
  theme_minimal()
```

```r
cleaned_df |>
  ggplot(aes(x = Glucose, y = BMI, color = as.factor(Outcome))) +
  geom_point(alpha = 0.7) +
  geom_smooth(method = "lm", se = FALSE) +
  labs(color = "Outcome", title = "Scatter Plot of Glucose vs BMI") +
  theme_minimal()
```

```r
cleaned_df |>
  pivot_longer(-Outcome, names_to = "Variable", values_to = "Value") |>
  ggplot(aes(x = as.factor(Outcome), y = Value, fill = as.factor(Outcome))) +
  geom_boxplot() +
  facet_wrap(~Variable, scales = "free") +
  labs(x = "Outcome", fill = "Outcome", title = "Boxplots by Outcome") +
  theme_minimal()
```

```r
# Calculate correlation matrix
cor_matrix <- cleaned_df |>
  select(-Outcome) |>
  cor(use = "complete.obs")

# Plot the heatmap
corrplot(cor_matrix, method = "circle", type = "upper", tl.col = "black", tl.srt = 45, title = "Correlation Matrix")
```

```r
#Outcome Proportion Bar Plot
cleaned_df |>
  group_by(Outcome) |>
  summarise(Count = n()) |>
  ggplot(aes(x = as.factor(Outcome), y = Count, fill = as.factor(Outcome))) +
  geom_bar(stat = "identity") +
  labs(x = "Outcome", y = "Count", fill = "Outcome", title = "Outcome Proportion") +
  theme_minimal()
```

```r
#Pairwise plot
cleaned_df |>
  select(Glucose, BMI, Age, Insulin, Outcome) |>
  ggpairs(aes(color = as.factor(Outcome), alpha = 0.7)) +
  theme_minimal()
```
